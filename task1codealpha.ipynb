```python
import time

name = input("What is you Name :  ")
print("hello,  "+name+"  time to play hangman!!")

print(" ")

time.sleep(1)

print("Start Processing.....")

time.sleep(0.5)

secret_word="elephant"
print("\nhint: This is a large,gray animal.")

guesses = ""

turns = 10

while turns > 0:
    failed = 0

    for char in secret_word:
        if char in guesses:
            print(char, end=" ")
        else:
            print("_", end=" ")
            failed += 1
    if failed == 0:
        print("\nYou've cracked the code!!")
        break

    print()

    guess = input("guess a charcter: ")
    guesses += guess

    if guess not in secret_word:
        turns -= 1
        print("\nOops, that's not quite right. ")
        print("you have", turns, "more guesses")

    if turns == 0:
        print("\nUh-oh, you have run out of tries!")
        print("The word was: ", secret_word)
```
